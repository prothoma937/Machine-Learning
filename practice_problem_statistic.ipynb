```python
# A2. Manual Min-Max on a tiny set
# Given scores = [20, 25, 30, 50], scale to [0, 1] by hand. Show each step.
```

```python
import numpy as np
import pandas as pd
df=pd.DataFrame({'value':[20, 25, 30, 50]})
df
```

```python
min=np.min(df['value'])
max=np.max(df['value'])
range=max-min
min_max=(df['value']-min)/range
print(min_max)
```

```python
# A3. Z-scores on a subset
# Given x = [8, 9, 11], compute mean, standard deviation, then standardize each. Use population standard deviation for this question.
```

```python
mean=df['value'].mean()
std=df['value'].std()
z_score=(df['value']-mean)/std
print(z_score)
print(std)
```

```python
# A4. Robust scaling ingredients
# Given y = [5, 6, 6, 7, 50], find median, Q1, Q3, IQR. Do not scale yet.
```

```python
median=df['value'].median()
q1=df['value'].quantile(0.25)
q3=df['value'].quantile(0.75)
iqr=q3-q1
r_cal=(df['value']-median)/iqr
r_cal=r_cal.round(2)
print(r_cal)
```

```python
# B5. Vectors and alignment [Optional]
#  a = [3, −1, 2], b = [4, 0, −2], c = [−6, 2, −4]
#  Tasks:
#  a) Compute a·b and a·c
#  b) Compare signs and magnitudes to comment on the alignment of a with b and with c
#  c) L2 normalize a and give the normalized vector to three decimals
```

```python
a = np.array([3, -1, 2])
b = np.array([4, 0, -2])
c = np.array([-6, 2, -4])
dot_product1=a@b #partially aligned
dot_product2=a@c #anti-aligned
print(dot_product1,dot_product2)
```

```python
#  c) L2 normalize a and give the normalized vector to three decimals
l2=np.linalg.norm(a)
normalized_a =np.round(a/l2,3)
print(l2, normalized_a)
```
