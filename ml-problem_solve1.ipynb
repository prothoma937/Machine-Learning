```python
import numpy as np
import pandas as pd
df=pd.DataFrame({'Salary':[45, 48, 50, 52, 55, 58, 60, 62, 65, 68, 70, 72, 75, 78, 80, 85, 90, 95, 150, 500]})
df.head()
```

```python
salary_range = df['Salary'].max() - df['Salary'].min()
salary_range
```

```python
s_var=np.var(df['Salary'])
s_std=np.std(df['Salary'])

q1=df['Salary'].quantile(0.25)
q2=df['Salary'].quantile(0.50)
q3=df['Salary'].quantile(0.75)

IQR=q3-q1
lower=q1-1.0*IQR
upper=q1+1.0*IQR
print(IQR,lower,upper)
outlier=df[(df['Salary']<lower) | (df['Salary']>upper)]
print('outlier:',outlier)
```
